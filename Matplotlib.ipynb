```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(car_sales)
plt.show()
```

```python
sns.histplot(data = car_sales, x =  'Price_in_thousands', hue = 'Vehicle_type')
plt.show
# Vehicle_type and their Price
```

```python
sns.boxplot(data = car_sales, x = 'Sales_in_thousands', hue = '__year_resale_value')
plt.show
#Yearly sales and their resale value
```

```python
sns.catplot(data = car_sales, x = 'Fuel_efficiency')
plt.show
#fuel efficiency of a car
```

```python
car_sales.plot(x = 'Vehicle_type')
# vehicle_type and its associated with the factors
```

```python
sns.countplot(data = car_sales, x = 'Manufacturer')
sns. set(rc={'figure.figsize':(60,15)})
plt.show
#Number of Cars Manufactured by a company
```

```python
sns.histplot(data = car_sales, x = 'Power_perf_factor')
sns. set(rc={'figure.figsize':(12,15)})
#Power_perf_factor of a car
```

```python
sns.scatterplot(data = car_sales, x = 'Model', y = 'Fuel_capacity')
sns. set(rc={'figure.figsize':(60,30)})
plt.show
#Fuel capacity according to the model of a car
```
